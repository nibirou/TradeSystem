# 一个“统一数据预处理模块”的可落地架构，把你现有的：

数据载入dataload、因子计算factorengine、打标签labelbuilder

- DataLoad（多股票池、日频/5min、交易日历、buffer 窗口裁剪、去重）
- FactorEngine（量价因子 + 可插拔扩展 + 基本面融合 AkShare/baostock 季频）
- LabelBuilder（支持日频 close 版本，也支持 5min 未来窗口内 best close / 止损路径版）
- 数据清洗与处理（缺失、重复、winsorize/zscore、中性化接口、对齐交易日网格）

一套统一的量化数据载入模块化代码

提示词prompt：我目前做了好几个量化策略，各自的数据载入dataload、因子计算factorengine、打标签labelbuilder这些数据预处理模块不是很统一，我现在想设计统一的数据预处理模块。我给你看我的几个策略各自的数据载入模块以及调用方式，我希望你能帮我整合到一起，如果设计的数据载入代码过长，你可以分模块输出。保留我给你代码具备的，所有功能（比如因子插拔），在这个前提下尽可能的全面优化，我的要求就是数据载入灵活支持多数据源、多类型数据（量价、基本面、算法挖掘出的新因子、定量处理后的金融文本等），带上一些数据处理（缺失值处理、重复值处理、中性化处理等）；因子计算灵活可插拔、支持随时插入新因子、支持自定义因子选取的开始结束时间及时间窗口间隔（日频、周频、月频以及自定义时间窗口）；对应因子选取，打标签labelbuilder支持 价格、收益率、波动率在自定义开始结束时间及时间窗口间隔下的计算，并且 价格、收益率、波动率 可以支持用日频收盘价统计，或者基于5分钟量价数据统计（比如计算收益率时，取未来n日内5min线的最优close价作为卖出价）。除此之外可能还有一些我没有写到，但你认为可能有必要的功能、或者需要注意的点，请你一并考虑。
